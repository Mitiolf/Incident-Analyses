```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

np.random.seed(42)

n = 5000

priorities = ["Low", "Medium", "High"]

categories = [
    "Application",
    "Database",
    "Network",
    "Security",
    "Access"
]

teams = [
    "App_Team",
    "DB_Team",
    "Infra_Team",
    "Security_Team",
    "Support_Team"
]

start_date = datetime(2024,1,1)

data = []

for i in range(n):

    created = start_date + timedelta(
        minutes=random.randint(0, 60*24*180)
    )

    priority = random.choices(
        priorities,
        weights=[0.5,0.35,0.15]
    )[0]

    category = random.choices(categories, weights=[0.5,0.15,0.15,0.10,0.10])[0]

    team = random.choice(teams)

    # Time resolution depends on priority
    if priority == "High":
        resolution_minutes = int(np.random.exponential(scale=120))
    elif priority == "Medium":
        resolution_minutes = int(np.random.exponential(scale=360))
    else:
        resolution_minutes = int(np.random.exponential(scale=720))

    resolved = created + timedelta(minutes=resolution_minutes)

    # SLA depends on priority
    if priority == "High" and resolution_minutes < 15:
        sla_breach = 0
    elif priority == "Medium" and resolution_minutes < 60:
        sla_breach = 0
    elif priority == "Low" and resolution_minutes < 600:
        sla_breach = 0
    else:
        sla_breach = 1


    reopen_count = np.random.poisson(0.3)

    data.append([
        i+1,
        created,
        resolved,
        priority,
        category,
        team,
        sla_breach,
        reopen_count,
        resolution_minutes
    ])

df = pd.DataFrame(data, columns=[
    "incident_id",
    "created_date",
    "resolved_date",
    "priority",
    "category",
    "assigned_team",
    "sla_breached",
    "reopen_count",
    "resolution_minutes"
])

df.to_csv("incidents.csv", index=False)


df.head()
```
